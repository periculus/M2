# M2 Jupyter Kernel Magic Commands Demo

This notebook demonstrates all the magic commands available in the M2 Jupyter kernel.

```macaulay2
-- First, let's see all available magic commands
%help
```

## 1. Logging Control

By default, M2 process logs are saved to `~/.m2_kernel_logs/`. You can disable this for cleaner operation:

```macaulay2
-- Check current logging status
%logging
```

```macaulay2
-- Disable logging for cleaner operation
%logging off
```

## 2. LaTeX Display Control

```macaulay2
-- Create a ring and ideal for testing
R = QQ[x,y,z,w]
I = ideal(x^3 + y^3 + z^3 + w^3 - 1, x^2*y + y^2*z + z^2*w + w^2*x)
```

```macaulay2
-- Default: LaTeX is enabled
matrix{{1,2,3},{4,5,6},{7,8,9}}
```

```macaulay2
-- Disable LaTeX for plain text output
%latex off
```

```macaulay2
-- Now output is in plain text
matrix{{1,2,3},{4,5,6},{7,8,9}}
```

```macaulay2
-- Re-enable LaTeX
%latex on
```

## 3. Font Size Control

```macaulay2
-- Default size
det matrix{{x,y},{z,w}}
```

```macaulay2
-- Make it larger
%fontsize=Large
```

```macaulay2
-- Now the output is larger
det matrix{{x,y},{z,w}}
```

```macaulay2
-- Try small size
%fontsize=small
det matrix{{x,y},{z,w}}
```

```macaulay2
-- Reset to normal
%fontsize=normalsize
```

## 4. Timeout Management

For long computations, you can adjust the timeout:

```macaulay2
-- Check current timeout
%timeout
```

```macaulay2
-- Set a longer timeout for complex computations
%timeout=600
```

```macaulay2
-- Now this computation has 10 minutes to complete
decompose I
```

## 5. Debug Mode

```macaulay2
-- Enable debug mode for detailed logging
%debug on
```

```macaulay2
-- Disable debug mode
%debug off
```

## 6. Using Output References (oN)

Notice how each output is labeled with `o1`, `o2`, etc. You can reference these in later cells:

```macaulay2
-- Reference previous outputs
-- If I was output o2, we can use it:
gens o2
```

## Tips

1. Use `%logging off` for normal work to avoid filling up disk space
2. Hover over output to see the "Copy LaTeX" button
3. The heartbeat indicator (if installed) shows kernel activity in the status bar
4. All settings persist for the session
