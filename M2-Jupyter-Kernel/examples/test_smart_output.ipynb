# Smart Output Handling Tests

This notebook demonstrates the smart output handling feature that automatically decides whether to use LaTeX or plain text based on output size.

Thresholds:
- Matrices larger than 50x50 → plain text
- LaTeX content > 10KB → plain text
- Polynomials with > 150 terms → plain text

## Small/Medium Outputs (LaTeX Rendering Enabled)

```macaulay2
-- Small 3x3 matrix - should show LaTeX
matrix {{1,2,3},{4,5,6},{7,8,9}}
```

```macaulay2
-- Medium 25x25 matrix - should still show LaTeX
M = random(ZZ^25, ZZ^25)
```

```macaulay2
-- Small polynomial - should show LaTeX
R = QQ[x,y,z]
f = x^3 + y^3 + z^3 - 3*x*y*z
```

```macaulay2
-- Small ideal - should show LaTeX
I = ideal(x^2, x*y, y^2)
```

## Large Outputs (LaTeX Rendering Disabled)

For these outputs, you should see a blue notice: "⚡ LaTeX rendering disabled for large output. Showing plain text for better performance."

```macaulay2
-- Large 60x60 matrix - should disable LaTeX
BigM = random(ZZ^60, ZZ^60)
BigM
```

```macaulay2
-- Long polynomial with 200 terms - should disable LaTeX
S = QQ[t]
longPoly = sum for i from 1 to 200 list t^i
longPoly
```

```macaulay2
-- Large resolution - may disable LaTeX depending on size
T = QQ[a,b,c,d,e]
J = ideal(a^4, a^3*b, a^2*b^2, a*b^3, b^4, a^3*c, a^2*b*c, a*b^2*c, b^3*c, a^2*c^2, a*b*c^2, b^2*c^2, a*c^3, b*c^3, c^4)
res J
```
