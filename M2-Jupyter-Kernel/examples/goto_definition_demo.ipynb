# Go-to-Definition Demo

This notebook demonstrates the go-to-definition feature in the M2 Jupyter kernel.

## How to use:
- **Alt+Click** on any symbol to jump to its definition
- **Shift+Hover** to see quick documentation
- **Tab** for autocompletion

```macaulay2
if true then "Sverrir"
```

```macaulay2
-- Define a polynomial ring
R = QQ[x,y,z]

-- Define an ideal
I = ideal(x^2 - y, y^2 - z, z^2 - x)

-- Define a function
mySquare := n -> n^2
```

Now try Alt+Clicking on the symbols below to jump to their definitions:

```macaulay2
-- Alt+Click on 'R' to see where it was defined
describe R

-- Alt+Click on 'I' to jump to its definition
gens I

-- Alt+Click on 'mySquare' to see the function definition
mySquare(5)
```

## Built-in Symbols

Go-to-definition also works for built-in M2 symbols:

```macaulay2
-- Alt+Click on 'gb' to see it's a built-in function
G = gb I

-- Alt+Click on 'ideal' or 'matrix' 
J = ideal matrix {{x,y},{y,z}}
```

## Complex Example

```macaulay2
-- Define multiple related objects
S = QQ[a,b,c,d]
f = map(R, S, {x^2, y^2, z^2, x*y*z})
ker f
```
