# Symbol Definition Magic Command Demo

Since go-to-definition with mouse clicks requires frontend support that may not be available,
we provide the `%def` (or `%where`) magic command to show where symbols are defined.

```macaulay2
-- First, let's define some objects
R = QQ[x,y,z]
I = ideal(x^2 - y, y^2 - z, z^2 - x)
```

```macaulay2
-- Now use %def to see where R was defined
%def R
```

```macaulay2
-- Check where I was defined
%where I
```

```macaulay2
-- Define a function
f = n -> if n <= 1 then 1 else n * f(n-1)
```

```macaulay2
-- Check the function definition
%def f
```

```macaulay2
-- Built-in symbols show as built-in
%def gb
```

```macaulay2
-- Check a type
%def Ring
```

```macaulay2
-- Undefined symbols
%def undefinedSymbol
```

## Complex Example

```macaulay2
-- Define multiple related objects
S = QQ[a,b,c,d]
phi = map(R, S, {x^2, y^2, z^2, x*y*z})
K = ker phi
```

```macaulay2
if true then print "sverrir"
```

```macaulay2
-- Check multiple definitions
%def S
print "---"
%def phi
print "---"
%def K
```

## Notes

- The `%def` command tracks all symbols defined in the current session
- Built-in M2 symbols (functions, types, constants) are recognized
- The command shows:
  - File location (notebook cells show as `<notebook>`)
  - Line number within the cell
  - Cell ID
  - Type (variable, function, method)
  - The actual code that defined the symbol
