## Ring Theory

```macaulay2
-- Create a polynomial ring
R = QQ[x,y,z]
```

```macaulay2
-- Define some polynomials
f = x^2 + y^2 - 1
g = x*y - z^2
```

```macaulay2
-- Create an ideal
I = ideal(f, g)
```

## Linear Algebra

```macaulay2
-- Create a matrix
M = matrix {{1, 2, 3}, {4, 5, 6}, {7, 8, 9}}
```

```macaulay2
-- Compute determinant
det M
```

```macaulay2
-- Random matrix over finite field
S = GF(7)[a,b]
N = random(S^3, S^3)
```
