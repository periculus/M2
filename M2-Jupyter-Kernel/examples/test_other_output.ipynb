# M2 'Other Output' Display Test

This notebook demonstrates the new feature that displays informational output from M2 (comments, progress messages) in a separate light blue block.

## Operations that produce 'other output'

Many M2 operations produce informational messages that were previously filtered out. Now they're displayed in a light blue block above the main output.

```macaulay2
-- Resolution computation often produces informational messages
R = QQ[x,y,z]
I = ideal(x^2, x*y, y^2, x*z, y*z, z^2)
res I
```

```macaulay2
-- Gröbner basis computation with trace
gbTrace = 1
J = ideal(x^3 - y, y^3 - z, z^3 - x)
gb J
```

```macaulay2
-- Primary decomposition can produce progress messages
K = ideal(x*y, x*z, y*z)
decompose K
```

## Comparison with %pi magic

When using %pi magic, the 'other output' is captured and displayed differently (as a flowing blue banner).

```macaulay2
-- Without %pi - other output in light blue block
gbTrace = 2
gb J
```

```macaulay2
-- With %pi - other output captured by progress tracker
%pi 2 gb J
```

## Multiple commands in one cell

Each command's 'other output' is displayed with its corresponding result.

```macaulay2
-- Multiple operations
gbTrace = 1
L = ideal(x^2 - y, y^2 - z)
gb L
res L
```

## Visual styling

The 'other output' block has:
- Light blue background (#f0f8ff)
- Slightly darker blue left border
- Monospace font
- Subtle opacity (0.9)

This makes it visually distinct from the main output while still being readable.
