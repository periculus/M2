# Line Magic Semantics Test

This notebook demonstrates that line magic (`%pi`) only affects M2 code on the same line.

```macaulay2
-- Setup
R = QQ[x,y,z]
I = ideal(x^2, y^2, z^2)
J = ideal(x^3, y^3, z^3)
```

## Test 1: Line magic with code on same line
Only `gb J` should have progress tracking

```macaulay2
%pi 2 gb J
res I
```

## Test 2: Line magic without code on same line
No progress tracking - magic has no effect

```macaulay2
%pi 2
gb J
```

## Test 3: Multiple line magics in one cell
Each magic only affects its own line

```macaulay2
%pi 1 gb I
%pi 2 gb J
res I
```

## Test 4: Cell magic affects all statements

```macaulay2
%%pi 2
gb I
gb J
res I
```
