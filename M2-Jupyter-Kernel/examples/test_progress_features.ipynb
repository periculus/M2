# M2 Progress Indicators Demo

This notebook demonstrates the progress tracking features for long-running M2 computations.

```macaulay2
-- Setup a ring for testing
R = QQ[x,y,z,w]
```

```macaulay2
-- Create a moderately complex ideal
I = ideal(x^3 + y^3 + z^3 + w^3 - 1, x^2*y + y^2*z + z^2*w + w^2*x, x*y*z + y*z*w + z*w*x + w*x*y)
```

```macaulay2
-- Compute Gröbner basis (should show progress)
gb I
```

```macaulay2
-- Compute a resolution (should show progress)
res I
```

```macaulay2
-- Test with a larger polynomial ring
S = QQ[a,b,c,d,e,f,g,h]

J = ideal(a^2-b*c, b^2-c*d, c^2-d*e, d^2-e*f, e^2-f*g, f^2-g*h, g^2-h*a, h^2-a*b)
```

```macaulay2
o10
```

```macaulay2
-- This should trigger progress indicators
%pi 2
gb J
```

```macaulay2
%pi 2 gb J
```

```macaulay2
%pi 2
gb J
```
